# Remaking Default ROBERT Graphs
This is a notebook to remake the default ROBERT graphs after training predictive models. This script will output SVG files of the graphs. Required input is data from completed ROBERT testing runs in the correct folders

```python
# importing packages
import matplotlib.pyplot as plt
import glob 
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
# key colors
paton_dark_blue = '#000080'
paton_teal_blue = '#376e8f'
paton_skyish_blue = '#529cbe'
paton_pale_blue = '#9dc2d9'
paton_dark_green = '#5c8625'
paton_light_green = '#c2d9a5'
paton_pale_green = '#d9e7c6'

paton_bright_red = '#fa0004'
paton_maroon = '#98002d'
paton_hot_pink = '#d00d44'
paton_pale_pink = '#f9cbd5'
paton_grey1 = '#b9b0a6'
paton_grey2 = '#d8d0c7'
paton_grey3 = '#f0ece7'

paton_blue = '#4865a2'
paton_dark_grey = '#808080'
paton_teal = '#127b8a'
paton_light_grey = '#bebebe'
```

```python
aqme_path = '../Data/ROBERT_Results/Semi-Empirical/'
dft_path = '../Data/ROBERT_Results/DFT'

aqme_best_model = 'MVL_85'
dft_best_model = 'MVL_85'
```

```python
# parody plot for Semi-Empirical model
ml_train = pd.read_csv(f'{aqme_path}/PREDICT/{aqme_best_model}_train_No_PFI.csv')
ml_valid = pd.read_csv(f'{aqme_path}/PREDICT/{aqme_best_model}_valid_No_PFI.csv')
ml_test = pd.read_csv(f'{aqme_path}/PREDICT/{aqme_best_model}_test_No_PFI.csv')

ml_train_data = ml_train[['dG_C5','dG_C5_pred']]
ml_valid_data = ml_valid[['dG_C5', 'dG_C5_pred']]
ml_test_data = ml_test[['dG_C5', 'dG_C5_pred']]

# ml_train

plt.plot([-150,-110],[-150,-110], color=paton_light_grey, linestyle='--', linewidth=2)
plt.scatter(x=ml_train_data['dG_C5'],y=ml_train_data['dG_C5_pred'],
           color=paton_dark_blue, label='Train', s=75, marker='o',alpha=0.5,edgecolors=paton_dark_blue)
plt.scatter(x=ml_valid_data['dG_C5'],y=ml_valid_data['dG_C5_pred'],
           color=paton_teal, label='Valid', s=75, marker='D',alpha=0.75,edgecolors=paton_teal)
plt.scatter(x=ml_test_data['dG_C5'],y=ml_test_data['dG_C5_pred'],
           color=paton_pale_blue, label='Test', s=75, marker='s',alpha=0.9,edgecolors=paton_pale_blue)

# plt.title('Predictions of GB_80_PFI', fontsize=20)
plt.title('AQME Best Model PFI', fontsize=20)

plt.xlabel(r'Calculated $\Delta G$', fontsize=20)
plt.xticks(np.arange(-145,-119,5),fontsize=15)
plt.xlim(-146,-119)

plt.ylabel(r'Predicted $\Delta G$', fontsize=20)
plt.yticks(np.arange(-145,-119,5),fontsize=15)
plt.ylim(-146,-119)

plt.legend(fontsize=15)

# plt.savefig(f'robert_aqme_parody_{aqme_best_model}_c5_12descp_update.svg', dpi=300, bbox_inches='tight')
```

```python
# parody plot for DFT
ml_train = pd.read_csv(f'{dft_path}/PREDICT/{dft_best_model}_train_No_PFI.csv')
ml_valid = pd.read_csv(f'{dft_path}/PREDICT/{dft_best_model}_valid_No_PFI.csv')
ml_test = pd.read_csv(f'{dft_path}/PREDICT/{dft_best_model}_test_No_PFI.csv')

ml_train_data = ml_train[['dG_C5','dG_C5_pred']]
ml_valid_data = ml_valid[['dG_C5', 'dG_C5_pred']]
ml_test_data = ml_test[['dG_C5', 'dG_C5_pred']]

# ml_train

plt.plot([-150,-110],[-150,-110], color=paton_light_grey, linestyle='--', linewidth=2)
plt.scatter(x=ml_train_data['dG_C5'],y=ml_train_data['dG_C5_pred'],
           color=paton_dark_blue, label='Train', s=75, marker='o',alpha=0.5,edgecolors=paton_dark_blue)
plt.scatter(x=ml_valid_data['dG_C5'],y=ml_valid_data['dG_C5_pred'],
           color=paton_teal, label='Valid', s=75, marker='D',alpha=0.75,edgecolors=paton_teal)
plt.scatter(x=ml_test_data['dG_C5'],y=ml_test_data['dG_C5_pred'],
           color=paton_pale_blue, label='Test', s=75, marker='s',alpha=0.9,edgecolors=paton_pale_blue)

# plt.title('Predictions of GB_80_PFI', fontsize=20)
plt.title('DFT Best Model PFI', fontsize=20)

plt.xlabel(r'Calculated $\Delta G$', fontsize=20)
plt.xticks(np.arange(-145,-119,5),fontsize=15)
plt.xlim(-146,-119)

plt.ylabel(r'Predicted $\Delta G$', fontsize=20)
plt.yticks(np.arange(-145,-119,5),fontsize=15)
plt.ylim(-146,-119)

plt.legend(fontsize=15)

# plt.savefig(f'robert_dft_parody_{dft_best_model}_c5_8descp_update.svg', dpi=300, bbox_inches='tight')
```

```python
## Heatmap data for DFT descriptors for dG C5

dft_heatmap_path = '../Data/ROBERT_Results/DFT/full_dataset_model_determination'

gb_60_pfi = pd.read_csv(f'{dft_heatmap_path}/GENERATE/Raw_data/PFI/GB_60_PFI.csv')
gb_65_pfi = pd.read_csv(f'{dft_heatmap_path}/GENERATE/Raw_data/PFI/GB_65_PFI.csv')
gb_70_pfi = pd.read_csv(f'{dft_heatmap_path}/GENERATE/Raw_data/PFI/GB_70_PFI.csv')
gb_75_pfi = pd.read_csv(f'{dft_heatmap_path}/GENERATE/Raw_data/PFI/GB_75_PFI.csv')
gb_80_pfi = pd.read_csv(f'{dft_heatmap_path}/GENERATE/Raw_data/PFI/GB_80_PFI.csv')
gb_85_pfi = pd.read_csv(f'{dft_heatmap_path}/GENERATE/Raw_data/PFI/GB_85_PFI.csv')
gb_90_pfi = pd.read_csv(f'{dft_heatmap_path}/GENERATE/Raw_data/PFI/GB_90_PFI.csv')

mvl_60_pfi = pd.read_csv(f'{dft_heatmap_path}/GENERATE/Raw_data/PFI/MVL_60_PFI.csv')
mvl_65_pfi = pd.read_csv(f'{dft_heatmap_path}/GENERATE/Raw_data/PFI/MVL_65_PFI.csv')
mvl_70_pfi = pd.read_csv(f'{dft_heatmap_path}/GENERATE/Raw_data/PFI/MVL_70_PFI.csv')
mvl_75_pfi = pd.read_csv(f'{dft_heatmap_path}/GENERATE/Raw_data/PFI/MVL_75_PFI.csv')
mvl_80_pfi = pd.read_csv(f'{dft_heatmap_path}/GENERATE/Raw_data/PFI/MVL_80_PFI.csv')
mvl_85_pfi = pd.read_csv(f'{dft_heatmap_path}/GENERATE/Raw_data/PFI/MVL_85_PFI.csv')
mvl_90_pfi = pd.read_csv(f'{dft_heatmap_path}/GENERATE/Raw_data/PFI/MVL_90_PFI.csv')

nn_60_pfi = pd.read_csv(f'{dft_heatmap_path}/GENERATE/Raw_data/PFI/NN_60_PFI.csv')
nn_65_pfi = pd.read_csv(f'{dft_heatmap_path}/GENERATE/Raw_data/PFI/NN_65_PFI.csv')
nn_70_pfi = pd.read_csv(f'{dft_heatmap_path}/GENERATE/Raw_data/PFI/NN_70_PFI.csv')
nn_75_pfi = pd.read_csv(f'{dft_heatmap_path}/GENERATE/Raw_data/PFI/NN_75_PFI.csv')
nn_80_pfi = pd.read_csv(f'{dft_heatmap_path}/GENERATE/Raw_data/PFI/NN_80_PFI.csv')
nn_85_pfi = pd.read_csv(f'{dft_heatmap_path}/GENERATE/Raw_data/PFI/NN_85_PFI.csv')
nn_90_pfi = pd.read_csv(f'{dft_heatmap_path}/GENERATE/Raw_data/PFI/NN_90_PFI.csv')

rf_60_pfi = pd.read_csv(f'{dft_heatmap_path}/GENERATE/Raw_data/PFI/RF_60_PFI.csv')
rf_65_pfi = pd.read_csv(f'{dft_heatmap_path}/GENERATE/Raw_data/PFI/RF_65_PFI.csv')
rf_70_pfi = pd.read_csv(f'{dft_heatmap_path}/GENERATE/Raw_data/PFI/RF_70_PFI.csv')
rf_75_pfi = pd.read_csv(f'{dft_heatmap_path}/GENERATE/Raw_data/PFI/RF_75_PFI.csv')
rf_80_pfi = pd.read_csv(f'{dft_heatmap_path}/GENERATE/Raw_data/PFI/RF_80_PFI.csv')
rf_85_pfi = pd.read_csv(f'{dft_heatmap_path}/GENERATE/Raw_data/PFI/RF_85_PFI.csv')
rf_90_pfi = pd.read_csv(f'{dft_heatmap_path}/GENERATE/Raw_data/PFI/RF_90_PFI.csv')

heatmap_data = pd.concat([gb_60_pfi,gb_65_pfi,gb_70_pfi,gb_75_pfi,gb_80_pfi,gb_85_pfi,gb_90_pfi,
                          mvl_60_pfi,mvl_65_pfi,mvl_70_pfi,mvl_75_pfi,mvl_80_pfi,mvl_85_pfi,mvl_90_pfi,
                          nn_60_pfi,nn_65_pfi,nn_70_pfi,nn_75_pfi,nn_80_pfi,nn_85_pfi,nn_90_pfi,
                          rf_60_pfi,rf_65_pfi,rf_70_pfi,rf_75_pfi,rf_80_pfi,rf_85_pfi,rf_90_pfi])

heatmap_data = heatmap_data[['train', 'model', 'rmse']]

# Replace 'MVL' with 'MLR' in the 'model' column
heatmap_data['model'] = heatmap_data['model'].replace({'MVL': 'MLR'})

sns.set(font_scale=1.2, style='ticks')

result = heatmap_data.pivot(index='train', columns='model', values='rmse')

ax = sns.heatmap(result, annot=True, fmt='.2f', cmap=sns.light_palette(paton_teal, as_cmap=True), vmin=0, 
                 vmax=10, linewidth=3, linecolor='white', cbar_kws={'label': 'RMSE Validation Set'})

# Flip the y-axis
ax.invert_yaxis()

plt.ylabel('Training Size (%)', fontsize=20)
plt.yticks(fontsize=15)
plt.xlabel('ML Model Type', fontsize=20)
plt.xticks(fontsize=15)
plt.title('Heatmap ML Models (PFI)', fontsize=20)

plt.tick_params(axis='both', which='major', labelsize=15)
ax.patch.set_edgecolor('black')
ax.patch.set_linewidth(5)

ax.figure.axes[-1].yaxis.label.set_size(20)
cax = ax.figure.axes[-1]
cax.tick_params(labelsize=15)

sns.despine(top=False, right=False)

# plt.savefig('dft_heatmap.svg', dpi=300, bbox_inches='tight')
```
